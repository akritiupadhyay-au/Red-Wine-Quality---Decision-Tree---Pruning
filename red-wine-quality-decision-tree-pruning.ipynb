### Identifying correlation among variables.

```python
corr = data.corr()
for i in range(len(corr)):
    print('Positive correlation :\n')
    a = ((corr.iloc[i]>0.5) & (corr.iloc[i]!=1))
    print(a[a==True],'\n\n')
    b = ((corr.iloc[i]<(-0.5))&(corr.iloc[i]!=1))
    print('Negative correlation :\n')
    print(b[b==True],'\n\n')
```

```python
data[features].plot(kind='box',subplots = True,layout = (5,5),figsize = (20,20))
plt.show()
```

```python
Outliers = []
for colname in features:
    col = data[colname]
    Q1 = np.quantile(col,0.25)
    Q3 = np.quantile(col,0.75)
    IQR = Q3-Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    for i in col:
        if (i<lower or i>upper):
            Outliers.append(i)
    print(colname)
    print(Outliers)
    for x in Outliers:
        data.drop(data[(data[colname]==x)].index,inplace = True)
    
```

```python
path = model.cost_complexity_pruning_path(x_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
print(ccp_alphas)
```

```python
models = []
for ccp_alpha in ccp_alphas:
    model = tree.DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    model.fit(x_train, y_train)
    models.append(model)
```

```python
train_score = np.array([i.score(x_train,y_train) for i in models])
test_score = np.array([i.score(x_test,y_test) for i in models])
score_df = pd.DataFrame(data = [models,train_score.T,test_score.T]).transpose()
score_df.columns = ['model','train score','test score']
score_df.head()
```
